```python
import pandas as pd
import numpy as np

df = pd.read_csv('marketing_campaign_data_messy.csv')
print(f"Loaded Dataset: {df.shape[0]} rows, {df.shape[1]} columns")
```

```python
df
```

```python
print(df.columns.tolist())
```

```python
# HEADER CLEANER

df.columns = df.columns.str.strip().str.lower().str.replace(' ', '_')
print(df.columns.tolist())
```

```python
# TYPE CONVERSION & CURRENCY CLEANING

clean_spend = df['spend'].astype(str).str.contains('$')
print(df.loc[clean_spend, ['campaign_id', 'spend']].head(5))

df['spend'] = df['spend'].astype(str).str.replace(r'[^\d.-]', '', regex=True)
df['spend'] = pd.to_numeric(df['spend'])

print('FIX APPLIED')
print(df.loc[clean_spend, ['campaign_id', 'spend']].head(5))
```

```python
# CATEGORICAL TYPOS

print(df['channel'].unique())

cleanupMap =  {
    'Facebok' : 'Facebook',
    'Insta_gram' : 'Instagram',
    'Tik_Tok' : 'TikTok',
    'Gogle' : 'Google Ads',
    'E-mail' : 'Email',
    'N/A' : np.nan
}
df['channel'] = df['channel'].replace(cleanupMap)
print('FIX APPLIED')
print(df['channel'].unique())
```

```python
# HANDLING MISSING VALUES

print(df['active'].unique())

boolMap = {
    'Yes' : True,
    'No' : False,
    'Y' : True,
    '0' : False,
    '1' : True,
    1 : True,
    0 : False
}
df['active'] = df['active'].map(boolMap).fillna(False).astype(bool)
print('FIX APPLIED')
print(df['active'].unique())
```

```python
# DATE PARSING

print(df['start_date'].dtypes)

df['start_date'] = pd.to_datetime(df['start_date'], errors='coerce')
df['end_date'] = pd.to_datetime(df['end_date'], dayfirst=True, errors='coerce')

print('FIX APPLIED')
print(df['start_date'].dtypes)
```

```python
df = df.loc[:, ~df.columns.duplicated()]
```

```python
df
```

```python
# LOGICAL INTEGRITY CHECKS (CLICKS SHOULD NOT EXCEED IMPRESSIONS)

invalid_clicks = df['clicks'] > df['impressions']
print(df.loc[invalid_clicks, ['campaign_id', 'clicks', 'impressions']].head(5))
```

```python
# LOGICAL INTEGRITY CHECKS (TIME TRAVEL)

invalid_dates = df['end_date'] < df['start_date']
print(df.loc[invalid_dates, ['campaign_id', 'start_date', 'end_date']].head(5))

df.loc[invalid_dates, 'end_date'] = df.loc[invalid_dates, 'start_date'] + pd.Timedelta(days=30)
print('FIX APPLIED')
print(df.loc[invalid_dates, ['campaign_id', 'start_date', 'end_date']].head(5))
```

```python
# HANDLING OUTLIERS

Q1 = df['spend'].quantile(0.25)
Q3 = df['spend'].quantile(0.75)

IQR = Q3 - Q1
upperLimit = Q3 + (3 * IQR)

outlierMask = df['spend'] > upperLimit
print(df.loc[outlierMask, ['campaign_id', 'spend']].head(5))

print('FIX APPLIED')
df.loc[outlierMask, 'spend'] = upperLimit
print(df.loc[outlierMask, ['campaign_id', 'spend']].head(5))
```

```python
# STRING PARSING (EXTRACTING DEVICE FROM CAMPAIGN NAME)

print(df['campaign_name'].head(5))

df['season'] = df['campaign_name'].str.extract(r'Q\d_([^_]+)')
print('EXTRACTED SEASON')
print(df[['campaign_name', 'season']].head(5))
```

```python
col = df.pop('season')
df.insert(2, 'season', col)
```
